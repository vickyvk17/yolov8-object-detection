```python
from ultralytics import YOLO
import cv2
import json
import os
```

```python
def detect_holes_and_export_json(model_path, image_folder, output_json_path):

    model = YOLO(model_path)

    results_dict = {}


    for img_name in os.listdir(image_folder):
        if not img_name.lower().endswith(('.png', '.jpg', '.jpeg')):
            continue

        img_path = os.path.join(image_folder, img_name)
        img = cv2.imread(img_path)


        results = model(img)


        detections = []
        for result in results:
            for box in result.boxes:

                class_id = int(box.cls)
                class_name = model.names[class_id]
                confidence = float(box.conf)
                x1, y1, x2, y2 = map(int, box.xyxy[0])


                center_x = (x1 + x2) / 2
                center_y = (y1 + y2) / 2

                detections.append({
                    "class": class_name,
                    "confidence": confidence,
                    "bbox": [x1, y1, x2, y2],
                    "center": [center_x, center_y]
                })


        hole_count = sum(1 for d in detections if d["class"] == "Holes")
        threaded_hole_count = sum(1 for d in detections if d["class"] == "Threaded_Holes")



        results_dict[img_name] = {
            "total_holes": hole_count,
            "total_threaded_holes": threaded_hole_count,
            "detections": detections
        }


    with open(output_json_path, 'w') as f:
        json.dump(results_dict, f, indent=4)

    print(f"Results saved to {output_json_path}")


detect_holes_and_export_json(
    model_path='/content/runs/detect/metal_holes_detection5/weights/best.pt',  # your trained model
    image_folder='/content/', # Assuming the image is in the content folder based on previous cells
    output_json_path='holes_detection_results.json'
)
```
